## Maior e Menor Valor

maior = max(lista)

menor = min(lista)

```python
vendas = [1000, 1500, 15000, 270, 900, 100, 1200]
```

- Qual o item mais vendido?
- Qual o item menos vendido?

```python
mais_vendido = max(vendas)
print(mais_vendido)
menos_vendido = min(vendas)
print(menos_vendido)
```

```python
i = vendas.index(mais_vendido)
nome_produto_mais_vendido = produtos[i]
print(nome_produto_mais_vendido)
i = vendas.index(menos_vendido)
nome_produto_menos_vendido = produtos[i]
print(nome_produto_menos_vendido)
```
