## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]
vendas_1sem.extend(vendas_2sem)
maior_valor_mes = max(vendas_1sem)
menor_valor_mes = min(vendas_1sem)
i = vendas_1sem.index(maior_valor_mes)
ii = vendas_1sem.index(menor_valor_mes)
print(f'Melhor mês de vendas do ano é {meses[i]} e o menor mês é {meses[ii]}')
print(vendas_1sem)

```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
print(f'O melhor mês do ano foi {meses[i]} com {vendas_1sem[i]} vendas')
print(f'O pior mês do ano foi {meses[ii]} com {vendas_1sem[ii]} vendas')
print(f'Faturamento total do ano é {sum(vendas_1sem)} e o faturamento do melhor mês repesentou {vendas_1sem[i] / sum(vendas_1sem):.1%}')
```

## 3. Crie uma lista com o top 3 valores de vendas do ano (sem fazer "no olho")

Dica: o método remove retira um item da lista.

```python
top3 = []
vendas_1sem.sort()
for l in vendas_1sem[0:9]:
    vendas_1sem.remove(l)
    print(vendas_1sem)
    if len(vendas_1sem) == 3:
        top3 += vendas_1sem
        print(f'top 3 valores de vendas do ano {top3}')

```
